```python
import pandas as pd
```

```python
visitation = pd.read_csv('visitation.csv')
```

```python
badlands = visitation.loc[visitation['unit_name'] == 'Badlands National Park']
```

```python
badlands = badlands.iloc[1:, :]
badlands = badlands[::-1]
badlands = badlands.rename(columns={"yearraw": "year"})
badlands.plot('year', 'visitors', figsize=(20, 5))
```

```python
import pandas as pd
```

```python
visitation = pd.read_csv('visitation.csv')
```

```python
tetons = visitation.loc[visitation['unit_name'] == 'Grand Teton National Park']
```

```python
tetons = tetons.iloc[1:,:]
tetons = tetons[::-1]
tetons.plot('yearraw', 'visitors', figsize=(20, 5))
```
